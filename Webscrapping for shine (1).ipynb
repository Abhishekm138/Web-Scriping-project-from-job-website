```python
import pandas as pd
import numpy as np
import time
import requests
import selenium
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from bs4 import BeautifulSoup
import urllib3
```

```python
def getUrl(key):
    key = key.replace(' ','-')
    temp = 'https://www.shine.com/job-search/{}-jobs'
    url = temp.format(key)
    return url
```

```python
url = getUrl('python')
print(url)
```

```python
res = requests.get(url)
print(res)
```

```python
# driver = webdriver.Chrome()
# driver.get(url)
```

```python
soup = BeautifulSoup(res.content,'html.parser')
soup
```

```python
cards = soup.find_all('div', class_='jobCard_jobCard__jjUmu white-box-border jobCard')
num_cards = len(cards)
print(num_cards)
```

```python
card = cards[0]
card
```

```python
# Fetching job_name
job_name = card.find('a').text if card.find('a') else 'Not Mentioned'
print(job_name)
```

```python
# Fetching company_location
company_name_element = card.find('div', 'jobCard_jobCard_cName__mYnow')
if company_name_element:
    company_name = company_name_element.text
else:
    company_name = 'Not Mentioned'
print(company_name)
```

```python
# Fetching experience_required
card_element = card.find('div', 'jobCard_jobCard_lists_item__YxRkV jobCard_jobIcon__3FB1t')
if card_element:
    experience_required = card_element.text
else:
    experience_required = 'Not Mentioned'
print(experience_required)
```

```python
# Fetching job_description
job_card = card.find('ul', 'jobCard_jobCard_jobDetail__jD82J')
if job_card:
    job_description = job_card.text
else:
    job_description = 'Not Mentioned'
print(job_description)
```

```python
# Fetching job_apply_url
card_link = card.find('a')
if card_link:
    job_apply_url = 'https://www.shine.com' + card_link['href']
else:
    job_apply_url = 'Not Mentioned'

print(job_apply_url)
```

```python
def find_job(card):
    # Fetching job_name
    job_name = card.find('a').text if card.find('a') else 'Not Mentioned'
    
    # Fetching company_name
    company_name = card.find('div','jobCard_jobCard_cName__mYnow').text if card.find('div','jobCard_jobCard_cName__mYnow') else 'Not Mentioned'
   
    # Fetching company_location
    company_location = card.find('div','jobCard_jobCard_lists_item__YxRkV jobCard_locationIcon__zrWt2').text if card.find('div','jobCard_jobCard_lists_item__YxRkV jobCard_locationIcon__zrWt2') else 'Not Mentioned'
    
    # Fetching experience_required
    experience_required = card.find('div','jobCard_jobCard_lists_item__YxRkV jobCard_jobIcon__3FB1t').text if card.find('div','jobCard_jobCard_lists_item__YxRkV jobCard_jobIcon__3FB1t') else 'Not Mentioned'
    
    # Fetching job_description
    job_description = card.find('ul','jobCard_jobCard_jobDetail__jD82J').text if card.find('ul','jobCard_jobCard_jobDetail__jD82J') else 'Not Mentioned'
    
    # Fetching job_apply_url
    job_apply_url = 'https://www.shine.com' + card.find('a')['href'] if card.find('a') and 'href' in card.find('a') else 'Not Mentioned'
    
    all_job_detail = (job_name, company_name, company_location, experience_required, job_description, job_apply_url)
    return all_job_detail
```

```python
all_jobs = []
for i in cards:
    temp = find_job(i)
    all_jobs.append(temp)
```
